```julia
# plot a histogram of NFKB2 expression in ABC and GC DLBCL lines
using StatsPlots
using DataFrames
using Query
using CSV
using HypothesisTests
using Statistics
```

```julia
#linesToHighlight=["U2932","SUDHL10","TMD8","RIVA","RCK8"]
cooFileName="nfkb2Histogram.csv"
cooDF = DataFrame(CSV.File(cooFileName))
cooDF2 = deepcopy(cooDF)
cooDF2=filter(:COO => n -> n == "GCB", cooDF2)
cooDF=filter(:COO => n -> n == "ABC", cooDF)
plotly()
allGeneNames=propertynames(cooDF)[9:end]
for gene in allGeneNames
    
    
    thisPlot=boxplot([1], cooDF2[!,gene], label="ABC",linewidth=2)
    #@df singers_moscow violin!(string.(:VoicePart), :Height, side=:left, linewidth=0, label="Moscow")
    dotplot!([1], cooDF2[!,gene], marker=(:blue,stroke(0)), label="",mode = :density)

    boxplot!([2], cooDF[!,gene], label="GC",linewidth=2)
    #@df singers_moscow violin!(string.(:VoicePart), :Height, side=:left, linewidth=0, label="Moscow")
    dotplot!([2], cooDF[!,gene], marker=(:green,stroke(0)), label="",ylim=(-2,5),xlim=(0,4),ylabel=String(gene)*" - Counts Per Million (log10)",mode = :density)
    #@df singers_moscow dotplot!(string.(:VoicePart), :Height, side=:left, marker=(:black,stroke(0)), label="")
    xticks!([1,2], [""],size=(300,600))
    pval=pvalue(ApproximateTwoSampleKSTest(cooDF[!,gene], cooDF2[!,gene]); tail = :both)
    annotate!(1.5,3.75,"p="*string(round(pval,digits=3)))

#     thisPlot=violin([1], cooDF2[!,gene], linewidth=0, label="GC",side=:left)
#     #@df singers_moscow violin!(string.(:VoicePart), :Height, side=:left, linewidth=0, label="Moscow")
#     dotplot!([1], cooDF2[!,gene], side=:left, marker=(:blue,stroke(0)), label="",mode = :density)

#     violin!([1], cooDF[!,gene], linewidth=0, label="ABC",side=:right)
#     #@df singers_moscow violin!(string.(:VoicePart), :Height, side=:left, linewidth=0, label="Moscow")
#     dotplot!([1], cooDF[!,gene], side=:right, marker=(:green,stroke(0)), label="",xlim=(0,2),ylim=(0.0,3.5),ylabel=String(gene)*" - Counts Per Million (log10)",mode = :density)
#     #@df singers_moscow dotplot!(string.(:VoicePart), :Height, side=:left, marker=(:black,stroke(0)), label="")
#     xticks!([1], [""])
    display(thisPlot)
end
```
